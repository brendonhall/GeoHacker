# Unboxing Core Photos

This notebook demonstrates how to extract slabbed core photographs from the image templates often produced by service companies.  Core images from the Pharos-1 well in the Poseidon dataset are used in this example.  The Poseidon dataset is licensed [CC-BY 3.0](https://creativecommons.org/licenses/by/3.0/au/) by Geoscience Australia and ConocoPhillips.

Use these links to download the Pharos-1 core images from the Poseidon [Google Drive](https://drive.google.com/drive/folders/1s7y9mFMZN9YgsqXSw2WnyOYrNwqUoEbc) link:
- [White Light Photos](https://drive.google.com/drive/folders/1tLrVugETP4NRHiQQr-dZ5-zAzgVAX6Xq?usp=sharing)
- [UV Photos](https://drive.google.com/drive/folders/1ZqS0OXc2qgdao7-mSusBP96WZxEQEncC?usp=sharing)
- [CT Images](https://drive.google.com/drive/folders/1To8x6n0S75nC8W6tKr31KCsGlwofcjA_?usp=sharing)

```python
%matplotlib inline
import numpy as np 
from skimage import data as skdata
from skimage import io as skio
import matplotlib.pylab as plt
import glob
import os
#import imageio
from scipy.misc import imresize
from skimage.transform import rescale
from skimage import color
```

## White light images
Extract the white light images from the link above and set the `wl_image_dir` to the directory containing the images.  Here I extracted the files to a directory called `data/raw/` in the same location as the notebook.  Also set a directory to contain the cropped white light images.

```python
wl_image_dir = "data/raw/pharos-1_wl_1m_core2_5027_00m_to_5028_00m/"
wl_output_dir = "data/cleaned/WL/"

# distance (in pixels from top of image to the top of the core)
top =   642
# distance from left edge of image to left of core
left = 630
#distance from top of image to the bottom of the core
bottom = 12520
#distance from left of image to right edge of core
right = 1570

for full_filename in glob.glob(wl_image_dir + '*.jpg'):
    
    #extract the filename
    filename = os.path.basename(full_filename)

    # extract top depth from filename
    top_depth = filename[21:28] 
    # extract bottom depth
    bottom_depth = filename[33:40]
    
    img = skdata.imread(full_filename)
    
    img_cropped = img[top:bottom, left:right]
    
    #reduce the resolution of the image
    img_rescaled = rescale(img_cropped, .3, multichannel=True, anti_aliasing=True, mode='constant')
    
    #give the file a new name, with top and bottom depth
    new_filename = top_depth+'-'+bottom_depth+'.jpg'
    
    skio.imsave(os.path.join(wl_output_dir, new_filename), img_rescaled)
    print('Processed {}'.format(new_filename))
```

## UV images

```python
uv_image_dir = "data/raw/pharos-1_uv_1m_core2_5022_00m_to_5023_00m/"
uv_output_dir = "data/cleaned/UV/"

# use the same box locations as WL above

for full_filename in glob.glob(uv_image_dir + '*.jpg'):
    
    #extract the filename
    filename = os.path.basename(full_filename)

    # extract top depth from filename
    top_depth = filename[21:28] 
    # extract bottom depth
    bottom_depth = filename[33:40]
    
    img = skdata.imread(full_filename)
    
    img_cropped = img[top:bottom, left:right]
    
    #reduce the resolution of the image
    img_rescaled = rescale(img_cropped, .3, multichannel=True, anti_aliasing=True, mode='constant')
    
    #give the file a new name, with top and bottom depth
    new_filename = top_depth+'-'+bottom_depth+'.jpg'
    
    skio.imsave(os.path.join(uv_output_dir, new_filename), img_rescaled)
    print('Processed {}'.format(new_filename))
```

## CT data

```python
ct_image_dir = "data/raw/CT/"
ct_output_dir = "data/cleaned/CT/"


top =   680
left = 470
bottom = 3100
right = 700

left2 = 1020
right2 = 1240

for full_filename in glob.glob(ct_image_dir + '*.jpg'):
    
    filename = os.path.basename(full_filename)
    top_depth = filename[22:29]
    bottom_depth = filename[30:37] 
   
    new_filename1 = 'CTXZ_'+top_depth+'-'+bottom_depth+'.jpg'
    new_filename2 = 'CTYZ_'+top_depth+'-'+bottom_depth+'.jpg'

    img = skdata.imread(full_filename)
    img_cropped = img[top:bottom, left:right]
    
    img2_cropped = img[top:bottom, left2:right2]
    
    skio.imsave(os.path.join(ct_output_dir, new_filename1), img_cropped)
    skio.imsave(os.path.join(ct_output_dir, new_filename2), img2_cropped)
    
    print('Processed {}'.format(new_filename1))
```

## Display the cores

```python
import matplotlib.ticker as mtick
for count, filename in enumerate(glob.glob('data/cleaned/WL/*.jpg')):
    print(filename)
    basename = os.path.basename(filename)
    top_depth = float(basename[:7].replace('_','.'))
    bottom_depth = float(basename[8:15].replace('_','.'))
    base, ext = os.path.splitext(basename)
    base = base.replace('_', '-')
    
    img = skdata.imread(filename)
    uv_img = skdata.imread(os.path.join('data/cleaned/UV/', basename))
    ctx_img = skdata.imread(os.path.join('data/cleaned/CT/', 'CTXZ_'+base+'.jpg'))
    cty_img = skdata.imread(os.path.join('data/cleaned/CT/', 'CTYZ_'+base+'.jpg'))
    
    pixels_y = img.shape[0]
    # all core sections are 1m in length
    pixel_length = 1.0/pixels_y * 1000 # in mm
    pixel_area = pixel_length * pixel_length # in mm^2
    nh, nw, nc = img.shape
    
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(8,12))
    ax1.imshow(img, extent = [0, nw*pixel_length/1000, top_depth+1, top_depth])
    ax1.set_title('WL')
    ax1.yaxis.set_major_formatter(mtick.FormatStrFormatter('%.2f'))
    ax2.imshow(uv_img)
    ax2.set_title('UV')
    ax3.imshow(ctx_img)
    ax3.set_title('CT XZ')
    ax4.imshow(cty_img)
    ax4.set_title('CT YZ')    
    
    ax2.set_xticks([])
    ax3.set_xticks([])
    ax4.set_xticks([])

    ax2.set_yticks([])
    ax3.set_yticks([])
    ax4.set_yticks([])
    
    plt.show()
    
    #only display first 10 cores for now
    if count == 10:
        break
```
